```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# data 
tesla_year = df_tesla['MONTH'] # x-axis
tesla_search = df_tesla['TSLA_WEB_SEARCH']
tesla_stock = df_tesla['TSLA_USD_CLOSE']



# plot
# fig, ax = plt.subplots(figsize=(15,10))
# plt.xticks(rotation=90)


plt.figure(figsize=(15,10))

ax1 = plt.subplot()
l1, = ax1.plot(tesla_year, tesla_search, color='red')



ax2 = ax1.twinx()
l2, = ax2.plot(tesla_year, tesla_stock, color='orange')




# chart name
plt.title("number of lego parts form 1949-2020 ")

# axis names 
plt.xlabel('year')
plt.ylabel('Stock-Search')




# plt(tesla_year  , color='green' ,linewidth=2.0, label="Lego")


# plt.legend() # name for each line 
plt.legend(["Tesla Search", "Tesla Stock"])
```
